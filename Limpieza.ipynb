**`Transformaciones`**: 

+ Generar campo **`id`**: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = **`as123`**)

+ Los valores nulos del campo rating deberán reemplazarse por el string “**`G`**” (corresponde al maturity rating: “general for all audiences”

+ De haber fechas, deberán tener el formato **`AAAA-mm-dd`**

+ Los campos de texto deberán estar en **minúsculas**, sin excepciones

+ El campo ***duration*** debe convertirse en dos campos: **`duration_int`** y **`duration_type`**. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

+ Generar campo **`id`**: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = **`as123`**)

```python
# Agregamos la columna "id" en cada dataframe
# Especificamos que no tenga duplicados y que se posicione en la columna 0

amazon.insert(0, "id", "a"+amazon.show_id, allow_duplicates=False)

disney.insert(0, "id", "d"+disney.show_id, allow_duplicates=False)

hulu.insert(0, "id", "h"+hulu.show_id, allow_duplicates=False)

netflix.insert(0, "id", "n"+netflix.show_id, allow_duplicates=False)
```

+ Los valores nulos del campo rating deberán reemplazarse por el string “**`G`**” (corresponde al maturity rating: “general for all audiences”)

```python
#Remplazamos valores nulos por "G" con el método fillna y agregamos un Inplace=True para impactar

amazon["rating"].fillna("G", inplace=True)

disney["rating"].fillna("G", inplace=True)

hulu["rating"].fillna("G", inplace=True)

netflix["rating"].fillna("G", inplace=True)

#Sumamos los null, para ver que se impactaron los cambios
print(amazon.rating.isna().sum())
print(disney.rating.isna().sum())
print(hulu.rating.isna().sum())
print(netflix.rating.isna().sum())
```

+ De haber fechas, deberán tener el formato **`AAAA-mm-dd`**

```python
#con el método to_datetime, convertimos a fecha el campo date_added

amazon["date_added"] = pd.to_datetime(amazon["date_added"])
disney["date_added"] = pd.to_datetime(disney["date_added"])
hulu["date_added"] = pd.to_datetime(hulu["date_added"])
netflix["date_added"] = pd.to_datetime(netflix["date_added"])
```

```python
#Validamos que se cambio el formato
disney.head(3)
```

+ Los campos de texto deberán estar en **minúsculas**, sin excepciones


```python
#Seleccionamos las columnas con texto a modificar
columnas = ["type", "title", "director", "cast", "country", "rating", "duration", "listed_in", "description"]

#Cambiamos el tipo de dato de algunas columnas como cast, que se encuentra en float

hulu['cast'] = hulu['cast'].astype(object)

#Recorremos cada columna del df para pasarlo a minúscula

for i in columnas:
    amazon[i] = amazon[i].str.lower()
    disney[i] = disney[i].str.lower()
    hulu[i] = hulu[i].str.lower()
    netflix[i] = netflix[i].str.lower()
```

```python
amazon.info()
```

El campo duration debe convertirse en dos campos: duration_int y duration_type. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

```python
# Dividimos la columna en todos los df

amazon[['duration_int','duration_type']]=amazon.duration.str.split(' ',expand=True)
disney[['duration_int','duration_type']]=disney.duration.str.split(' ',expand=True)
hulu[['duration_int','duration_type']]=hulu.duration.str.split(' ',expand=True)
netflix[['duration_int','duration_type']]=netflix.duration.str.split(' ',expand=True)

#Validamos los diferentes términos que quedaron en duration_type

amazon.duration_type.unique()
disney.duration_type.unique()
hulu.duration_type.unique()
netflix.duration_type.unique()

#Estandarizamos el término Seasons

amazon.duration_type.replace("seasons","season", inplace=True)
disney.duration_type.replace("seasons","season", inplace=True)
hulu.duration_type.replace("seasons","season", inplace=True)
netflix.duration_type.replace("seasons","season", inplace=True)

#convertimos duration_int en integer. Remplazamos nulos si hay y convertimos
amazon["duration_int"].fillna(0, inplace=True)
disney["duration_int"].fillna(0, inplace=True)
hulu["duration_int"].fillna(0, inplace=True)
netflix["duration_int"].fillna(0, inplace=True)

amazon['duration_int'] = amazon['duration_int'].astype(int)
disney['duration_int'] = disney['duration_int'].astype(int)
hulu['duration_int'] = hulu['duration_int'].astype(int)
netflix['duration_int'] = netflix['duration_int'].astype(int)
```

```python
# Crear una columna con el nombre de la plataforma
amazon['Platform'] = 'amazon'
disney['Platform'] = 'disney'
hulu['Platform'] = 'hulu'
netflix['Platform'] = 'netflix'
```

```python
#Por último unimos los datasets para importar solo 1 archivo

All_Platforms = pd.concat([amazon, disney, hulu, netflix])

All_Platforms.info()

# Reseteamos indices
All_Platforms.reset_index(inplace=True)
```

```python
#Eliminamos columnas innecesarias
All_Platforms.drop(['index'], axis=1, inplace=True)
All_Platforms.drop(['duration'], axis=1, inplace=True)
```
